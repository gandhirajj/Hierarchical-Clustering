```python
import pandas as pd
```

```python
data=pd.read_csv("/content/diabetes dataset.zip")
```

```python
data
```

```python
y=data.pop("Outcome")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,y,test_size=0.2)
```

```python
list_acc=[]
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
for feature in x_train.columns:
  x_train_feature=x_train[[feature]]
  x_test_feature=x_test[[feature]]
  model=LogisticRegression()
  model.fit(x_train_feature,y_train)
  pred=model.predict(x_test_feature)
  acc=accuracy_score(y_test,pred)
  list_acc.append(acc)
```

```python
list_acc
```

```python
acc_ser=pd.Series(list_acc)
acc_ser.index=x_train.columns
```

```python
acc_ser.sort_values(ascending=False)
```

```python
from sklearn.feature_selection import RFE
model1=RFE(estimator=model,n_features_to_select=5)
model1.fit(x_train,y_train)
```

```python
model1.support_
```

```python
x_train.columns[model1.support_]
```

```python
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold
model2=RFECV(estimator=model,step=4,cv=StratifiedKFold(),scoring='accuracy')
model2.fit(x_train,y_train)
```

```python
model2.support_
```

```python
x_train.columns[model2.support_]
```
